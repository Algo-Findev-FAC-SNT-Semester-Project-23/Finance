```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
#Stock chosen is Infosys
infy_df=yf.download('INFY')
print(infy_df)
```

```python
#3 months have approximately 66 market working days
infy_df['SMA']=infy_df['Close'].rolling(66).mean()
print(infy_df['SMA'])
```

```python
#3 months have approximately 66 market working days
infy_df['EMA']=infy_df['Close'].ewm(66).mean()
print(infy_df['EMA'])
```

```python
plt.plot(infy_df.index, infy_df['Close'], label='Close Price')
plt.plot(infy_df.index, infy_df['SMA'], label='SMA')
plt.plot(infy_df.index, infy_df['EMA'], label='EMA')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Infosys Moving Averages')
plt.legend()
plt.show()
```
