```python
import numpy as np
import yfinance as yf
import seaborn as sb
import pandas as pd
import matplotlib.pyplot as plt
```

```python
stock = "INFY"
data = yf.download(stock)
```

```python
#Calculating SMA
sma = data['Close'].rolling(window=30*3).mean()

#Calculating EMA
ema = data['Close'].ewm(span=30*3, adjust=False).mean()
```

```python
plt.plot(sma.index, sma, label='SMA')
plt.plot(ema.index, ema, label='EMA')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title(f'INFY Moving Averages')
plt.legend()
plt.grid(True)
plt.show()
```
