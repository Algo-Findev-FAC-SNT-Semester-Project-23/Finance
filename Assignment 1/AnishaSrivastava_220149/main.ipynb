```python
stock = "TATAMOTORS.NS"
data = yf.download(stock,start="2023-01-01", end="2023-04-01")
```

```python
#Calculating SMA
sma = data['Close'].rolling(window=10).mean()

#Calculating EMA
ema = data['Close'].ewm(span=10, adjust=False).mean()
```

```python
plt.figure(figsize=(9, 6))
plt.plot(sma.index, sma, label='SMA')
plt.plot(ema.index, ema, label='EMA')
plt.ylabel('Price')
plt.xlabel('Date')
plt.grid(True)
plt.title(f'TATAMOTORS Moving Averages')
plt.legend()
plt.show()
```
