```python
import numpy as np
import pandas as pd
import yfinance as yf
stock = "BHARTIARTL.NS"
data = yf.download(stock,'2013-05-01','2023-05-01')
```

```python
#Calculating Simple Moving Average
data['SMA'] = data['Close'].rolling(90).mean()
print(data['SMA'])
```

```python
#Calculating Exponential Moving Average
data['EMA'] = data['Close'].ewm(span=90, adjust=False).mean()
print(data['EMA'])
```
