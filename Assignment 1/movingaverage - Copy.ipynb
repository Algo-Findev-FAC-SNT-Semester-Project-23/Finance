```python
import pandas as pd
import yfinance as yf
from yahoofinancials import YahooFinancials
```

```python
Ddff=pd.read_csv("C:/Users/HP/Documents/Downloads/ITC.NS.csv")
```

```python
Ddff
```

```python
average = []
for i in range(len(Ddff)):
    mean = Ddff.loc[i:i+5, 'Adj Close'].mean()
    average.append(mean)
```

```python
average
```

```python

expaverage = []
for i in range(len(Ddff)):
    if i == 0:
        mean1 = Ddff['Adj Close'][i]
    else:
        mean1 = Ddff['Adj Close'][i] * (2/(1+7)) + expaverage[i-1] * (1 - 2/8)
    expaverage.append(mean1)
```

```python
expaverage
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(expaverage)
plt.xlabel('Days')
plt.ylabel('Exponential Moving Average')
plt.title('Exponential Moving Average Plot')
plt.show()
```

```python
plt.plot(average)
plt.xlabel('Days')
plt.ylabel(' Moving Average')
plt.title('Simple Moving Average Plot')
plt.show()
```

```python
plt.plot(expaverage, label='Exponential Moving Average')
plt.plot(average, label='Simple Moving Average')
plt.xlabel('Days')
plt.ylabel('Moving Average')
plt.title('Combined Moving Average Plot')
plt.legend()
plt.show()
```
