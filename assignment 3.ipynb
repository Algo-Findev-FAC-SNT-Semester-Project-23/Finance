```python
df = pd.DataFrame({'stock_1': stock_1.close, 'stock_2': stock_2.close}, index=stock_1.index)
df['price_ratio']=(df.stock_1)/df.stock_2
df
```

```python
plt.plot(df.index,df.price_ratio,color='grey')
plt.xlabel('date')
plt.ylabel('price ratio')
```

```python
def calculate_rsi(data, window=14):
    price_change = data.diff(1)

    up_prices = price_change.where(price_change > 0, 0)
    down_prices = price_change.where(price_change < 0, 0)

    avg_gain = up_prices.rolling(window).mean()
    avg_loss = down_prices.abs().rolling(window).mean()

    rsi = 100 - (100 / (1 + (avg_gain/avg_loss)))
    return rsi
```

```python
df['rsi_stock_1']=calculate_rsi(df.stock_1)
df['rsi_stock_2']=calculate_rsi(df.stock_2)
df
```

```python
plt.plot(df.index,df.rsi_stock_1,color='grey')
plt.axhline(30,linestyle='--',color='yellow')
plt.axhline(70,linestyle='--',color='yellow')

```

```python
plt.plot(df.index,df.rsi_stock_2,color='grey')
plt.axhline(30,linestyle='--',color='yellow')
plt.axhline(70,linestyle='--',color='yellow')
```

```python
df['rsi_ratio']=df.rsi_stock_1/df.rsi_stock_2
df
```

```python
high_threshold = 1.2
low_threshold = 0.8
rsi_ratio=df.rsi_ratio

temp=0

signals = []
for i in range(len(rsi_ratio)):
    if rsi_ratio[i] > high_threshold and temp!=-1:
        signals.append(-1)  # sell stock_1, buy stock_2
        temp=-1
    elif rsi_ratio[i] < low_threshold and temp!=1:
        signals.append(1)   # buy stock_1, sell stock_2
        temp=1
    else:
        signals.append(0)   # No trade

print(signals)
```

```python
invest=100000
num_of_stocks_1=0
num_of_stocks_2=0
in_hand=invest
returns=[]

for i in range(len(signals)):
    if signals[i]==1:
        in_hand+=num_of_stocks_2*stock_2.close[i]
        num_of_stocks_2=0
        temp=in_hand//stock_1.close[i]
        in_hand-=temp*stock_1.close[i]
        num_of_stocks_1+=temp

    if signals[i]==-1:
        in_hand+=num_of_stocks_1*stock_1.close[i]
        num_of_stocks_1=0
        temp=in_hand//stock_2.close[i]
        in_hand-=temp*stock_2.close[i]
        num_of_stocks_2+=temp

    returns.append(in_hand+stock_1.close[i]*num_of_stocks_1+stock_2.close[i]*num_of_stocks_2)
```

```python
final_returns=returns[-1]
print("Return on investing Rs. 100000 is:",final_returns)

percentage_returns=format((((final_returns)-(invest))/invest)*(100),".2f")
print("Percentage return in case of combined strategy is:",percentage_returns,"%")
```
